```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x=np.linspace(0,5,11)
print(x)
y=x**2
print(y)
```

```python
plt.plot(x,y)
plt.title("First Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
plt.subplot(2,2,1)

plt.subplot(2,2,2)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
plt.plot(x,y)
```

```python
fig=plt.figure(figsize=(10,8),dpi=100)
axis1=fig.add_axes([.5,.5,.5,1])
axis1.plot(x,y)
axis2=fig.add_axes([.1,.1,.3,.3])
axis2.plot(x**2,y**2)
```

```python
plt.scatter(x,y)
```

```python
import seaborn as sbn
import matplotlib.pyplot as plt

df=sbn.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
plt.subplot(1,2,1)
sbn.histplot(df['total_bill'],bins=50,kde=True)
plt.subplot(1,2,2)
sbn.histplot(df['tip'],bins=50,kde=True)
```

```python
sbn.jointplot(x=df['total_bill'],y=df['tip'],kind='reg')
```

```python
sbn.pairplot(df,hue='sex',palette='rainbow')
```

```python
sbn.rugplot(df)
```

```python
sbn.barplot(x=df['sex'],y=df['total_bill'],estimator=np.sum)
```

```python
sbn.countplot(x=df['sex'],hue=df['day'])
```

```python
sbn.boxplot(x=df['tip'],y=df['day'])
```

```python
sbn.violinplot(y=df['total_bill'],x=df['day'])
```

```python
sbn.stripplot(x=df['day'],y=df['total_bill'])
```

```python
sbn.swarmplot(y=df['total_bill'],x=df['day'])
```
