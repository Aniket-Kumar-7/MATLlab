```python
match_wins=df['match_winner'].value_counts()
type(match_wins)
```

```python
sbn.barplot(y=match_wins.index,x=match_wins.values,palette='rainbow')
plt.title('Most winner Team')
plt.xlabel('Matches')
plt.ylabel('Teams')
```

```python
sbn.countplot(x=df['toss_decision'],palette='rainbow')
plt.title('Toss Decision')
```

```python
sbn.boxplot(x=df['match_winner'],y=df['toss_winner'])
vs=df[df['toss_winner']==df['match_winner']]['match_winner'].count()
vs
percentage=(vs*100)/df.shape[0]
percentage.round(2)
```

```python
wicket_win=df[df['won_by']=='Wickets']
print(wicket_win.count()['match_id'])
run_win=df[df['won_by']=='Runs']
(run_win.count())['match_id']
sbn.countplot(x=df['won_by'])
```
