```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x=np.linspace(0,5,11)
print(x)
y=x**2
print(y)
```

```python
plt.plot(x,y)
plt.title("First Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
plt.subplot(2,2,1)

plt.subplot(2,2,2)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
plt.plot(x,y)
```

```python
fig=plt.figure(figsize=(10,8),dpi=100)
axis1=fig.add_axes([.5,.5,.5,1])
axis1.plot(x,y)
axis2=fig.add_axes([.1,.1,.3,.3])
axis2.plot(x**2,y**2)
```

```python
plt.scatter(x,y)
```

```python
fig=plt.figure()
plt.hist(y)
```
